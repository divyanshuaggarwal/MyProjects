```python
import matplotlib.pyplot as plt
import pandas as pnd
import numpy as np
```

```python
data=np.genfromtxt('0000000000002329_training_diabetes_x_y_train.csv',delimiter=',')
```

```python
print(type(data))
```

```python
print(data.shape)
```

```python
x_train=data[:,:-1]
y_train=data[:,-1].reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_train,y_train)
```

```python
x_test=np.genfromtxt('0000000000002329_test_diabetes_x_test.csv',delimiter=',')
print(x_test.shape)
x_test
```

```python
y_predict=regressor.predict(x_test)
```

```python
y_predict.tolist()

y_preds=[]

for i in y_predict:
    y_preds.append(i[0])
a = []
for i in range(len(y_preds)):
    a.append(round(y_preds[i],5))
a
```

```python
preds = {'predictions': a }
```

```python
preds
```

```python
import pandas as pd

df = pd.DataFrame(preds, columns= ['predictions'])
```

```python
df.to_csv('Predictions.csv',index = False)
```
